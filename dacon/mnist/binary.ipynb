```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Conv2D,MaxPooling2D,BatchNormalization,Flatten,Dropout,Input,Activation,Dense
from tensorflow.keras.models import Sequential,Model,load_model
from sklearn.decomposition import PCA
from sklearn.preprocessing import OneHotEncoder
from tensorflow.keras.callbacks import EarlyStopping,ModelCheckpoint
from sklearn.model_selection import train_test_split

for k in range(0,10):
    df = pd.read_csv("train.csv",index_col=[0])
    modelpath = './binary_models/{}_binary.h5'.format(k)
    es = EarlyStopping(monitor='val_loss',patience=50)
    cp = ModelCheckpoint(monitor = 'val_loss',filepath=modelpath)
    
    label = df['digit']==k
    label = label.values.astype('float32')
    df['Label']=label
    x = df.values[:,2:-1].astype('float32')/255.0
    y = df.values[:,-1].astype('float32')
    # print(df)
    x_train,x_val,y_train,y_val = train_test_split(x,y,test_size=0.10,stratify=y)
    # print(x_train.shape,x_val.shape,y_train.shape,y_val.shape)
    # print(type(x_train),type(x_val),type(y_train),type(y_val))
    x_train = x_train.reshape(-1,28,28,1)#[:,2:26,2:26,:]
    x_val = x_val.reshape(-1,28,28,1)#[:,2:26,2:26,:]
    model = modeling_binary()
    model.compile(loss = 'binary_crossentropy',optimizer = 'adam',metrics=['accuracy'])
    model.fit(x_train,y_train,validation_data = (x_val,y_val),epochs=200,callbacks=[cp,es],batch_size=4)

# y = df.values[:,0]
# x = df.values[:,2:].astype('float32')/255.0#.reshape(-1,28,28)
# print(x.shape,y.shape)
```

```python
def modeling_binary():
    inputs = Input(shape=(28,28,1))
    x = inputs
    _x = Conv2D(256,3,padding='same')(x)
    _x = BatchNormalization()(_x)
    _x = Activation('relu')(_x)
    _x = Conv2D(128,3,padding='same')(_x)
    _x = BatchNormalization()(_x)
    _x = Activation('relu')(_x)
    x = _x
    _x = Conv2D(256,3,padding='same')(x)
    _x = BatchNormalization()(_x)
    _x = Activation('relu')(_x)
    _x = Conv2D(128,3,padding='same')(_x)
    _x = BatchNormalization()(_x)
    _x = Activation('relu')(_x)
    x = x+_x
    x = MaxPooling2D(2)(x)
    _x = Conv2D(512,3,padding='same')(x)
    _x = BatchNormalization()(_x)
    _x = Activation('relu')(_x)
    _x = Conv2D(128,3,padding='same')(_x)
    _x = BatchNormalization()(_x)
    _x = Activation('relu')(_x)
    x = x+_x
    _x = Conv2D(1024,3,padding='same')(x)
    _x = BatchNormalization()(_x)
    _x = Activation('relu')(_x)
    _x = Conv2D(128,3,padding='same')(_x)
    _x = BatchNormalization()(_x)
    _x = Activation('relu')(_x)
    x = x+_x
    x = MaxPooling2D(2)(x)
    x = Flatten()(x)
    x = Dense(1024)(x)
    x = Dense(1,activation='sigmoid')(x)
    outputs=x
    model = Model(inputs=inputs,outputs=outputs)
    return model
```
