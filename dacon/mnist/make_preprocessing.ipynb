```python
import pandas as pd
import numpy as np
def make_preprocessing():
    df2 = pd.read_csv("train.csv",index_col=[0])
    for k in range(80,170):
        print("임계값 : {} 실행중".format(k))
        df = pd.read_csv("train.csv",index_col=[0])
        for i in range(len(df2)):
            for j in range(784):
                temp = df2.iloc[i,j+2]
                if temp<k:
                    df.iloc[i,j+2]=0
        df.to_csv('preprocessing/preprocessing_{}.csv'.format(k))
```

```python
make_preprocessing()
```
